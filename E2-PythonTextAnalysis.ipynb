# Exercise 2.1

Complete the function that calculates the similarities of all texts in the dataframe

```python
import pandas as pd
import numpy as np
from difflib import SequenceMatcher
```

```python
data = pd.read_csv('E2-Data.csv', sep=';', encoding='ISO-8859-1').dropna()
data
```

```python
def similarity(data):
    
    #Comparing two strings, and considering blanks to be "junk"
    results = [round(SequenceMatcher(lambda x: x == " ", i, j).ratio(),2) for i in data['Text'] for j in data['Text']]
    #Convert the result to a Pandas Dataframe
    results = pd.DataFrame(np.array(results).reshape(data.shape[0],data.shape[0]), index=data.index, columns=data.index)
    
    assert results.shape == (data.shape[0], data.shape[0])
    
    return results
```

```python
#Run the function to find similarities between the text in the dataframe
similarity(data)
```

# Exercise 2.2

Calculate descriptive statistics of the dataset

- Pivot tables analyzing companies, universities, age and master semester.

```python
print('Número de estudiantes por carrera y semestre')
data.pivot_table(values='Age', index='Undergrad',columns='Semester', aggfunc='count')
```

```python
print('Min, Max y Promedio de edad por semestre')
data.pivot_table(values='Age', index='Semester',  aggfunc=[min, max, np.mean])
```

```python
print('Número de estudiantes por empresa y semestre')
data.pivot_table(values='Age', index='Company',columns='Semester', aggfunc='count')
```
